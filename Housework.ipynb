```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


file_path = r"D:\\EXCELWORKS\\House.xls"
data = pd.read_excel(file_path)


X = data["Size(sqft)"].values 
y = data["Price"].values  

def locally_weighted_regression(x_query, X, y, tau):
    m = len(X)
    weights = np.exp(-((X - x_query) ** 2) / (2 * tau ** 2))  
    
    X_b = np.c_[np.ones(m), X] 
    W = np.diag(weights)  
    theta = np.linalg.pinv(X_b.T @ W @ X_b) @ (X_b.T @ W @ y)
    x_query_b = np.array([1, x_query])
    return x_query_b @ theta
tau = 300  
x_query = 1600
predicted_price = locally_weighted_regression(x_query, X, y, tau)
print(f"Predicted Price for a 1600 sqft house: ${predicted_price:.2f}")


X_range = np.linspace(X.min(), X.max(), 100)
y_pred = np.array([locally_weighted_regression(x, X, y, tau) for x in X_range])

plt.scatter(X, y, color='red', alpha=0.5, label="Actual Data")
plt.plot(X_range, y_pred, color='blue', label="Locally Weighted Regression")
plt.scatter([x_query], [predicted_price], color='green', marker='o', label="Prediction (1600 sqft)")
plt.xlabel("House Size (sqft)")
plt.ylabel("Price (dollars)")
plt.legend()
plt.title("Locally Weighted Regression on House Prices")
plt.show()
```
